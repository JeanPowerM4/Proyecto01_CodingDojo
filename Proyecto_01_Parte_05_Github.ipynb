# Preprocesamiento

```python
y=df['Item_Outlet_Sales']
X=df.drop(columns=["Item_Outlet_Sales"])
```

```python
X.info()
X
```

```python
from sklearn.compose import make_column_selector
from sklearn.preprocessing import StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder
```

```python
num_selector = make_column_selector(dtype_include='number')
colums_nominal = ['Item_Fat_Content','Item_Type','Outlet_Type']
colums_ordinal = ['Outlet_Size','Outlet_Location_Type']
```

```python
standard_scaler = StandardScaler()
ordinal_encode = OrdinalEncoder()
nominal_encode = OneHotEncoder()
```

```python
col_transformer = make_column_transformer((make_pipeline(nominal_encode), colums_nominal),
                                          (make_pipeline(ordinal_encode), colums_ordinal),
                                          (make_pipeline(standard_scaler), num_selector),
                        remainder = 'passthrough')

col_transformer
```

```python
col_transformer.fit(X)
```

```python
aux = col_transformer.transform(X)
```

```python
aux=col_transformer.transform(X)
X_procesado=pd.DataFrame(aux,columns=col_transformer.get_feature_names_out())
X_procesado.info()
```

# KNN

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score

linear_reg = LinearRegression()
linear_reg.fit(X_train, y_train)

rf_reg = RandomForestRegressor()
rf_reg.fit(X_train, y_train)

knn_reg = KNeighborsRegressor()
knn_reg.fit(X_train, y_train)

# Predecir sobre los datos de prueba
y_pred_linear = linear_reg.predict(X_test)
y_pred_rf = rf_reg.predict(X_test)
y_pred_knn = knn_reg.predict(X_test)

# Evaluar los modelos
mse_linear = mean_squared_error(y_test, y_pred_linear)
r2_linear = r2_score(y_test, y_pred_linear)
print("Regresión Lineal:")
print("Error cuadrático medio (MSE):", mse_linear)
print("Coeficiente de determinación (R^2):", r2_linear)

mse_rf = mean_squared_error(y_test, y_pred_rf)
r2_rf = r2_score(y_test, y_pred_rf)
print("Bosque Aleatorio:")
print("Error cuadrático medio (MSE):", mse_rf)
print("Coeficiente de determinación (R^2):", r2_rf)

mse_knn = mean_squared_error(y_test, y_pred_knn)
r2_knn = r2_score(y_test, y_pred_knn)
print("KNN:")
print("Error cuadrático medio (MSE):", mse_knn)
print("Coeficiente de determinación (R^2):", r2_knn)
```
